## Import Data

```python
import pandas as pd
import numpy as np

df = pd.read_csv('data-latihan-kotor.csv')

df
```

```python
# Analisisi Dasar

print(df.head())
print(df.tail())
print(df.info())
print(df.describe())
```

```python
# Data Cleaning

## Mengisi Missing Value
df["kota"] = df["kota"].fillna(df["kota"].iloc[0])
df["harga"] = df["harga"].fillna(df["harga"].mean())

## Membenarkan Format Tanggal
df["tanggal_pembelian"] = pd.to_datetime(
    df["tanggal_pembelian"],
    format="mixed",
    dayfirst=True
)

## Membenarkan Hasil Yang Minus
df = df[df["jumlah"] > 0]

## Menyamakan Format Kota
df["kota"] = df["kota"].str.title().str.strip()
```

```python
## Membenarkan Format Harga

pd.options.display.float_format = '{:,.0f}'.format
df
```

```python
# Feature Engineering

## Membuat Kolom Revenue
df["revenue"] = df["harga"] * df["jumlah"]

## Membuat Kolom Bulan
df["bulan"] = df["tanggal_pembelian"].dt.month_name()

## Membuat Kolom Tahun
df["tahun"] = df["tanggal_pembelian"].dt.year

## Membuat Kolom Hari
df["hari"] = df["tanggal_pembelian"].dt.day_name()

df.to_csv("data-latihan-bersih", index=False)
```
